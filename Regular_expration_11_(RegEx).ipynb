# Regular expression -
            - Regular Express  powerful tool for matching pattern in a string they allow to search,match and manipulate string based on specific patterns.
            - Meta Character - like \d,\d+,\w,\w+,\D,\D+. Use to extract data What type of data we want.

1. re.match( pattern,string ) - Tries to Match a pattern from the beginning of the the string.

2. re.search(pattern,string) - Search the string for 1st occurrence off the pattern return a match object if found or None otherwise.pahle jo milega use hi print karega cahe vo midil me hi kyu na ho.

3. re.split(pritten,string) - He split the string by occurrence of the pattern and return a list of substring.

4. re.findall(pattern,string) - Use to find given value in Given string or Data.
5. re.sub(replace_value,) - Meanse substitute use to change or replace value .

```python
import re
```

# Match

```python
x = 'Machine learning is a sumset of AI is'
pattern = re.match("Machine",x)          # "Machinee" None return karge matlab exact match  hona chahiye. or search start se karega.

print(pattern)
```

```python
x = 'Machine learning is a sumset of AI'
pattern = re.match("Machine lea",x)
print(pattern.span())
print(pattern.start())
print(pattern.end())
```

# Search

```python
x = 'Machine learning is a sumset of AI is'
pattern = re.search('is',x)               # Search bss vo pahle wala hi string dega.
print(pattern)
```

# Subsitute

```python
x = 'Machine learning is a sumset of AI'
pattern = re.sub("is",'are',x)            # Subtitute

print(pattern)
```

```python
x = 'Machine learning is a sumset of AI'
pattern = re.sub("is",'are',x)            # Subtitute

print(pattern)
```

# Findall 

```python
x = 'Machine learning is a sumset of AI and Ai is booming'
pattern = re.findall("is",x)            # Find into the variable.

print(pattern)
```

# Use of \B, \b -
                - Use to find word.

```python
import re

text = "cat scatter category cat"   
# Match 'cat' as a whole word.
matches = re.findall(r"\bcat\b", text)
print(matches)
```

```python
import re

text = "scatter category"

# Match 'cat' only inside other words
matches = re.findall(r"\Bcat\B", text)
print(matches)
```

## 2. Use of ^ - 
                - To find pattern in the starting.

```python
text = 'hello Aastha you are  Looking Beauteful 22 Year old do you have any dog Ypur Id 856940'
pattern_1 =  r'^he.'                                             # \d use for digites.
pattern_2 = r'^y.u'
result_1 = re.findall(pattern_1,text)
result_2 = re.findall(pattern_2,text)
print(result_1)
print(result_2)
```

## 3. Use of * -
                - To find multiple pattern in the same string .

```python
text = 'hallo Aastha you are 22 Year old do you ah have any dog 78 Ypur hight Id 856940779'
pattern_1 =  r'ha*'                                          
pattern_2 =  r'78*'                           # Bas 7 hona chahiye.
pattern_3 = r'7$'
result_1 = re.findall(pattern_1,text)
result_2 = re.findall(pattern_2,text)
result_3 = re.findall(pattern_3,text)

print(result_1)
print(result_2)
print(result_3)
```

```python
text = 'hallo Aastha you are 22 Year old do you have any dog Ypur Id 856940'
pattern_1 =  r'^ha*'                                             
pattern_2 = r'^yo*'
result_1 = re.findall(pattern_1,text)
result_2 = re.findall(pattern_2,text)
print(result_1)
print(result_2)
```

# Types of Brackets

### 1. [ ]

```python
text = 'hallo Aastha you are 22 Year old do you have any dog 78 Ypur Id 1385694077'
pattern = r'[0-3]'
result = re.findall(pattern,text)
print(result)
```

```python
text = 'hallo Aastha you are 22 Year old do you have any dog 78 Ypur Id 1385694077'
pattern = r'[^0-3]'                     # Agar [] andar ^ to not ka kaam karega .
result = re.findall(pattern,text)
print(result)
```

```python
text = 'hallo Aastha you are 22 Year old do you have any dog 78 Ypur Id 1385694077'
pattern = r'[^0-3]+'                     # Agar [] andar ^ to not ka kaam karega . matlab 0-3 ko chod krr
result = re.findall(pattern,text)
print(result)
```

```python
text = 'hallo Aastha don  you are 22 Year old do you have any dog 78 Ypur Id 1385694077'
pattern = r'[a-e]'                    
result = re.findall(pattern,text)
print(result)
```

```python
text = 'hallo Aastha don  you are 22 Year old do you have any dog 78 Ypur Id 1385694077'
pattern = r'[^a-f]'                    
result = re.findall(pattern,text)
print(result)
```

### 2. {}

```python
import re

text = "hallo 2 3 63 Year 444 7777 dog"
pattern = r"[0-9]{2}"   # Exactly 2 digits , jada hai to cahlega kam nahi hona chahiye.
result = re.findall(pattern, text)
print(result)
```

```python
pattern = r"[0-9]{2,3}"   # for 2 to 3 digites only.
result = re.findall(pattern, text)
print(result)
```
