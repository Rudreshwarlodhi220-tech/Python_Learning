## Key , Value , Get , Item

```python
my_dict = {'Name':['Seeta',"Mahdav",'Manav'], 
          'Age':[23,24,52], 
          'Add':['B12','B13','B14']}
my_dict.keys()
```

```python
my_dict.values()
```

```python
my_dict.items() # Return values with Keys.
```

```python
for i,j in my_dict.items():
    print('Key',i,'Values',j)
```

```python
my_dict.get('Age')  # Get is a Built-in method.
```

```python
my_dict['Age']
```

## Add method

```python
a = {'a':1, 'b':2}
b = {'c':3, 'd':4}
a.update(b)

print(a)
```

```python
num={1:'One',2:'Two',3:'Three'}

print(num)
num.update({3:'THREE'})

print(num)

num.update({4:'four'})
print(num)
```

```python
my_dict = {'Name':['Seeta',"Mahdav",'Manav'], 
          'Age':[23,24,52], 
          'Add':['B12','B13','B14']}
copy_my_dict = my_dict.copy()

print('Origanal -\t',my_dict)
print('Copy -\t',copy_my_dict)
```

```python
copy_my_dict.update({'Customer':['Ram','Shyam','Ganesh']})
print(f'Origanal -\t {my_dict}\n')
print('Copy -\t',copy_my_dict)
```

```python
copy_my_dict['Customer'].append("HariOm")
print(copy_my_dict)
```

## Popitem

```python
num={1:'One',2:'Two',3:'Three'}

print(num)
new_num=num.pop(2)

print(num)
print(new_num)
```

## Clear

```python
a = {'a':1, 'b':2}
a.clear()
print(a)
```

```python
num={1:'One',2:'Two',3:'Three'}

print(num)
num.clear()  # Ye bass daata ko hata ta hai like variable or dictonury rahegi.

print(num)
```

```python
num={1:'One',2:'Two',3:'Three'}
del num # Pura hi haata deta hai 
# print(num)  # error throw karega. kyuki kuch hai hi nahi.
```

```python
num={1:'One',2:'Two',3:'Three'}

print('P1',num)
print('P2',num[2])
# print(num.get())  argue must

print('P3',num.get(3))
print('P4',num.get(7))  #  in this never throw error.

print('P5',num.keys())
print("\n'P6',",num.values())

print("\nP7",num.items())
```

```python
for i in num:
    print(i)
    print("\n",i,num[i])
```

```python
for i in num.values():
    print(i)
```

```python
num={1:'One',2:'Two',1:'Three',(3.0,4,5,):"four,five"} #[3,4,5,] not alowed also {3,4,5}
num[2]="TWO"
print(num)
```
