# Recursion - 
             - It use to call  function itself inside the function it's called recursion.

        # Recursive case   else:

```python
def factorial(n):
    if n <= 0:   # base case me last condition likhte hai . matlab kaha ja krr fun ke bahar hona chaiye.
        return 1
    else:
        return n*factorial(n-1) # recurcive case  
factorial(5)
```

# Lambda Function -
- As known as annomious(gumname) function .
- use to data manupulation, data handaling .

```python
add = (lambda a,b: a+b)
print(add(5,6))

mul = (lambda a,b,c : a*b*c)
print(mul(8,6,9))
```

# Filter 

```python
lst = [2,3,4,5,6,6,7,8,9,90,22,44]
# even = []
even = list(filter(lambda a : a%2 == 0,lst))  # Yaha list() hi likhana pdega nahi to ...
                                             # Map condition me True or falase  return kareega.
print(even)
```

# Reduce 

```python
from functools import reduce # Import karna jaruri hai
number=[1,2,3,4,5]

num_4 = reduce(lambda x,y : x+y ,number) # reduce sum frist 2 element then 2 ,2,2..

print(num_4)
```

# Global & Local variable 

```python
x = 100     #"It's called Global variable because it is out of the function . 
def fun():
    x = 400  #"It's called Local variable because it is inside  the function {ye bss fun ke andar hi x = 400 rahega} .
    return x
print("It's called Local variable = ",fun())
print("It's called Global variable = ",x)
```

```python
x = 100      
def fun():
    global x  # Jab ham chahate hai ki local variable Golobal baan jayee.
    x = 400  
    return x
print("It's called Local variable = ",fun())
print("It's called Global variable = ",x)
```
