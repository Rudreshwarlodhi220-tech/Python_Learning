### 2. Keyword Argument - 


```python
def greater(a,b,c):

                        # Called Docstring.
    if a>b and b>c:
        print('a is greater ')
    elif b>a and b>c:
        print('b is greater ')
    else:
        print('c is greater ')
x=10
y=20
z=30

greater(b=y,c=z,a=x)  # Yaha ham sequence aage piche kar saskte hai. Magar Positional me allow nahi hai.
```

### 3. Default Argument -

```python
def greater(a,b,c=99):  # Iss Argument me  ham pahle hi Parameter me value pass kar dete hai.

    print(a,b,c)
    
z=30

#greater(x,y,z)
greater(40,78)
greater(10,20,c=z)
```


### 4. *args (Arbitrary argument)-

```python
def add(*arg):   # Jab  hame pata na ho ki kitane argument aa sakte hai.
    total = 0
    for i in arg:
        total += i 
    return total
add((5,8,9,7,5,6,2,2,9,))
```

```python
a,b,*c = 10,2,3,4,5,6,7,8,9  # a,*b,c = *a,b,c
print(a)
print(b)
print(c)
```

```python
def sum(*arg):
    total = 0
    for i in arg:
        total+=i
    return total
print(sum(1,2,3,8,9))
```

```python
print(sum(*[1,2,3,6,5,8]))  # Pahle ye list ko allow nahi krr raha tha * ab karega.
```

## Fabonacci series

```python
def faba(n):
    a = 0
    b = 1
    sum = 0
    for i in range(n):
        sum = a+b
        a = b
        b = sum
        print(b)
# print(a)
# print(b)
faba(8)
```

```python
def fabo(n):
    a = 0
    b = 1
    print(a)
    print(b)
    for i in range(n):
        a,b = b, a+b
        print(b)
        #return b  # Yaha return use nahi hoga , Return akk baar hi execute hotaa hai bhali loop me hoi
fabo(8)
```

# Factorial 

```python
def fact(n):
    facto =1 
    for i in range(1,n+1):
        facto *= i
    #print(facto)
    return facto
fact(5)
```

```python
def gcd(a,b):
    GCD = 1
    if a>b:
        n = a
    else:
        n = b
    while n>0:
        if a%n ==0 and b%n == 0:
            GCD = n
            break
        n-=1
    return GCD
gcd(30,20)
```
