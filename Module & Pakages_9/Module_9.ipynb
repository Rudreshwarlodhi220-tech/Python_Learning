# Floor & ceil

```python
print(math.floor(25.51))
print(math.floor(25.49))
print(math.floor(25.99))      # 
print(math.floor(25.1))
```

```python
print(math.ceil(25))
print(math.ceil(25.51))
print(math.ceil(25.49))
print(math.ceil(25.99))
print(math.ceil(25.1))
```

# Squar root

```python
print(math.sqrt(36))   #  √36 = 6
print(math.cbrt(6548))   # ∛27 = 3
print(math.floor(math.cbrt(6548)))
```

# Date Time Modual 

```python
import datetime
```

```python
x =  datetime.datetime.now()   # Modual
print(x)
```

```python
date = x.strftime('%d-%m-%y')
print(date)
```

```python
day_name = x.strftime('%a')
print(day_name)

day_name = x.strftime('%A')
print(day_name)

day_name = x.strftime('%b')
print(day_name)

day_name = x.strftime('%B')
print(day_name)

day_name = x.strftime('%H')
print(day_name)

day_name = x.strftime('%M')
print(day_name)

day_name = x.strftime('%S')
print(day_name)
```
