# Lef shift Right Shift

```python
x = 10

print(bin(x)) # To print number to binary.

print('Left shift:', x << 2) # Left shift   10 * 2^2 = 10*4 = 40 isme multiply hoga 2 ki power 2 se.
print('Right shift:', x >> 2) # Right shift  10 / 2^2 = 10/4 = 2 issme divide hoga 2 ki power 2 se.

# haamesa x * 2 se hi hoga left shift me aur x / 2 se hi hoga right shift me.
```

# Identity oparator-

```python
x = 10
y = 10
u = 9

print(id(x))
print(id(y))
print(id(u),'\n\n')

print(x is y )
print(u is y)
print(u is not x)
```
