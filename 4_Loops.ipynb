# Break Statement

```python
for i in range(1,20):
    if i == 10:
        break 
    print(i)
```

# Continue statement

```python
for i in range(1,20):
    if i == 10 or i == 15:
        continue 
    print(i)
```

# Pass Statement

```python
for i in range(2,34):
    pass
```
