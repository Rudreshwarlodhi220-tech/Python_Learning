- Question 1: Extract all email addresses from a given text.  
- Test String: "Contact us at info@example.com or support@company.org for 
assistance." 

```python
import re
email = "Contact us at info@example.com or support@company.org for assistance." 
pattern = re.findall(r'\w+@\w+.\w+',email)
print(pattern)    # Here regex fillters automatically in which word contain @ and (.) only that word allowed.
```

2.Find all valid URLs (http or https) in a string.  
- Test String: "Visit https://www.python.org or http://example.com for more 
info." 

```python
site = "Visit https://www.python.org or http://example.com for more info."
pattern = re.findall(r'\w+://\w+.\w+.\w+',site)
print(pattern)
```

6. Problem: Validate if a string is a valid IPv4 address. 
- Test String 1: "192.168.1.1" Test String 2: "256.0.0.1" 

```python
Test_1 = "192.168.1.1" 
Test_2 = "256.0.0.1" 

pattern_1 = r'^((25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)\.){3}'\
                       r'(25[0-5]|2[0-4]\d|1\d\d|[1-9]?\d)$'
print(pattern_1)



def is_valid_ip(ip):
    return bool(re.match(pattern_1,ip))
print(is_valid_ip(Test_1))
print(is_valid_ip(Test_2))
```

```python
print(bool(re.match(pattern_1, "192.168.1.1")))  # True
print(bool(re.findall(pattern_1, "256.0.0.1"))) 
```

7. Problem: Replace all occurrences of "Python" (case-insensitive) with "Java". 
- Test String: "Python is fun. python is powerful. PYTHON is versatile." 

```python
import re
string =  "Python is fun. python is powerful. PYTHON is versatile." 
pattern = re.sub(r'Python','Java',string,flags=re.I)                # flag ka use  match karna matlab python kisi bhi form me ho like {Python,PYTHON,python} repalce karna hai.
print(pattern)
```

8. Problem: Remove all leading and trailing whitespace from a string. 
- Test String: "Hello World " 

```python
text = "   Hello World   "
pattern = re.sub(r'^\s+','',text)

pattern = re.sub(r'\s+$','',text)

print(pattern)
```

```python
text = "   Hello World   "

result = text.rstrip()       # Method 1
result = result.lstrip()

print(result)

print(text.strip())  # Method 2
```

10. Problem: Capitalize the first letter of each word in a sentence.
- Test String "hello world, how are you?" 

```python
text =  "hello world, how are you?" 
pattern_1 = re.findall(r'\b[a-z]',text)

for i in pattern_1:
    text = text.replace(i,i.upper())

print(pattern_1)
print(text)
```
