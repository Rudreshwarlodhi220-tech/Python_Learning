# Pelindrome

```python


x = 'Harsh' 
if x == x[::-1]:  # bhali loop nahi hai magar check pure element se karega
    print('Pelindrome')  

else:
    print('Not Pelindrome')
```

```python
x = 'rudra'
rev = ''
for i in x:
    rev = i + rev
    print(rev)

if x == rev :
    print('Pelindrome')
else:
    print('Not Pelindrome')
```

```python

input = 'string are immutable'

for i in input.split():
    print(i[::-1],end = ' ')
```

```python
input = 'string are immutable data'
rev1 = ''
rep2 = ''
rep3 = ''
count = 0
for i in input:
    if i == ' ':
        count+=1   
    
    if count == 0:
        rev1 = i + rev1

    if count == 1 :
        rep2 = i+ rep2

    elif count == 2:
        rep3 = i+ rep3
    
    

print(rev1,end= ' ')
print(rep2,end= ' ')
print(rep3,end= ' ')
```
