# - Count the number of vowels in a string.

```python
word = 'an Elephant'
vowels = ['a','e','i','o','u']
count = 0
for i in word.lower():  # Convert to lowercase to catch both upper and lower case vowels
    for j in vowels:
        if i == j:
            count += 1  # Increment count when a vowel is found
            # No need to check other vowels once we found a match
            break
print(count)
```

```python
word = 'an Elephant'
vowels = ['a', 'e', 'i', 'o', 'u','r']
count = 0

# Convert the word to lowercase so we catch both upper and lower case vowels
lower_word = word.lower()
i = 0

while i < len(lower_word):
    j = 0
    while j < len(vowels):
        if vowels[j] == lower_word[i]:
            count += 1  # Increment count when a vowel is found
            break       # No need to check other vowels once we found a match
        j += 1
    i += 1

print(count)
```
