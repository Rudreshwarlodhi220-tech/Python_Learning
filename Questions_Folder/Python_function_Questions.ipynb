```python
def chk_valid_email(email):
    spe_char = ['@','!','#','$','%','&','*','+','-','/','=','?','^','_','{','|','}','~']
    domain = ['.com','.org','.in']
    length = 64
    
    # 1. Length check
    if len(email) > length:
        return "Invalid Email"
    
    # 2. '@' होना चाहिए और सिर्फ एक बार
    if email.count('@') != 1:
        return "Invalid Email"
    
    # 3. Domain check
    if not any(email.endswith(d) for d in domain):
        return "Invalid Email"
    
    # 4. पहला character special नहीं होना चाहिए
    if email[0] in spe_char:
        return "Invalid Email"
    
    # 5. Digits और special symbols का गलत use avoid करो
    for ch in email:
        if ch == ' ':
            return "Invalid Email"
    
    return "Valid Email"


# Example Run
email = input('Enter your email = ')
print(chk_valid_email(email))
```

34. Write a function to rotate elements in a list to the right by a given
number of positions.
- input [1,2,3,4,5]
- output[4,5,1,2,3]

```python
def rotate_right(lst, k):
    n = len(lst)
    k = k % n   # Ye isliye  ki pata chale rotation kitani bar chalani hai. hr 5 vi bar me list rotation vese ka vesa hi ho jata hai , isliye % then 2 matlab rotation 2 brr bss.
    
    for _ in range(k):
        last = lst.pop()   # last element nikalane ke liye
        lst.insert(0, last)  # usko front mein daalane ke liye.
    return lst

# Example
numbers = [1, 2, 3, 4, 5]
print(rotate_right(numbers, 2))  # Output: [4, 5, 1, 2, 3]
```

35. Create a function that calculates the n-th term of the Fibonacci
sequence using memoization.

```python
def fibonacci(n, memo={}):
    v = 0 
    # Base cases
    if n == 0:
        return 0
    elif n == 1:
        return 1
    
    # Check if already calculated
    if n in memo:
        return memo[n]
    
    # Otherwise calculate and store
    memo[n] = fibonacci(n-1, memo) + fibonacci(n-2, memo)
    
    return memo[n]
# Example: 11th Fibonacci term
print(f"F(11) = {fibonacci(6)}")
```
