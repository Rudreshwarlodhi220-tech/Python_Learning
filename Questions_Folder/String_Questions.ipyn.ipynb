35. Question: Find length of longest substring without repeating characters. 
- Input: "abcabcbb" 
- Output: 3 

```python
text = "abcabcbb"

seen = {}
start = 0
max_len = 0

for i in range(len(text)):   
    ch = text[i]
    if ch in seen and seen[ch] >= start:
        start = seen[ch] + 1   
    seen[ch] = i
    max_len = max(max_len, i - start + 1)

print(max_len)
```
