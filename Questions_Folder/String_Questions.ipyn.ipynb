1. Question: Count the number of vowels in a given string.
- Input: "Hello World"
- Output: 3

```python
string = 'Hello world'
vowels = ['a','e','i','o','u']
count = 0

for i in string:
    for j in vowels:
        if i == j:
            count +=1
print(count)
```

3. Question: Check if a string is a palindrome.
- Input: "madam"
Output: True

```python
strr = 'Race Car'
lower_str = strr.replace(" ",'').lower()
rev_str = ''

for i in lower_str:
    if i == ' ':
        continue
    rev_str = i + rev_str
print('Revese string =',rev_str)
if rev_str == lower_str:
    print('Yes String is Palindrome')
else:
    print('No String is Palindrome')
```

4. Question: Remove all spaces from a string.

```python
msg = "hello World "
Count = ''
for i in msg:
    if i == ' ':
        continue
    Count += i
print(Count)
```

6. Question: Find the frequency of each character in a string.
- Input: "apple"
Output: {'a':1, 'p':2, 'l':1, 'e':1}


```python
string = 'Find the frequency of each character in a string'
new_str = string.replace(' ','')
dic = {}

for i in new_str:
    count = 0
    dic[i] = count
    for j in new_str:
        if i == j:
            count+=1
            
    dic[i] = count
    
print(dic)
```

7. Question: Find all positions of a particular character in a string.
- Input: "programming", character: 'g'
Output: [3, 10

```python
Input = "programming"
position = []
for i in range(len(Input)):
    if Input[i]=='g':
        position.append(i+1)

print(position)
```

10) Question: Remove all duplicate characters from a string.
- Input: "programming"
- Output: "progamin"


```python
text = 'programming'
strr = ''

for i in text:
    duplicate = False
    for j in strr:          
        if i == j:
            duplicate = True
            break
    if not duplicate:        
        strr += i

print(strr)
```
