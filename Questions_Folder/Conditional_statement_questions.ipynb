- >20L 6%

```python
loan = int(input('Enter Loan amount = '))
credit_score = int(input('Entere your Credit score = '))

if loan < 500000 :
    interest_rate = 10
elif loan > 500000 and loan<2000000:
    interest_rate = 8
elif loan > 2000000 :
    interest_rate = 6

if credit_score > 750:
    interest_rate -= 1
elif credit_score < 600:
    interest_rate += 2

interest_amount = (interest_rate / 100) * loan

print("Loan Amount =", loan)
print("Credit Score =", credit_score)
print("Final Interest Rate (%) =", interest_rate)
print("Interest Amount =", round(interest_amount, 2))
```
