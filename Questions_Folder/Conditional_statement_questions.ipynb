39. Write a program to calculate water bill:
● 0–50 units → ₹2 per unit
● 51–150 units → ₹5 per unit
● Above 150 → ₹8 per unit

```python
water_unit = int(input('Enter your water unit = '))
bill = 0

if water_unit <= 50: 
    bill = water_unit * 2
    print('Your water bill is =',bill)
elif water_unit > 50 and water_unit <= 150:
    bill = (water_unit - 50)*5 
    print('Your water bill is =',bill)
elif water_unit > 150: 
    bill = (water_unit - 150)*8 + 100*5 + 50*2
    print('Your water bill is =',bill)
```

40. Write a program to calculate mobile recharge tax:
