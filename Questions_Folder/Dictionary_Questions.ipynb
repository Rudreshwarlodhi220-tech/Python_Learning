1. Create a dictionary from two lists 
- Input: keys = ['a','b','c'], values = [1,2,3] 
- Output: {'a': 1, 'b': 2, 'c': 3} 

```python
keys = ['a','b','c']
values = [1,2,7]
dic = {}

for i in range(len(keys)):
    dic[keys[i]] = values[i]

print(dic)
```

तुमने loop for i in range(len(keys)): लगाया है।
इसका मतलब है कि i की value होगी: 0, 1, 2 (क्योंकि len(keys) = 3)।

जब i = 0 → dic[keys[0]] = values[0] → 'a': 1

जब i = 1 → dic[keys[1]] = values[1] → 'b': 2

जब i = 2 → dic[keys[2]] = values[2] → 'c': 3

👉 इसलिए values[i] लिखा है ताकि हर key के साथ उसी index का value जुड़ सके।

```python
keys = ['a','b','c']
values = [1,2,3]

dic = {keys[i]: values[i] for i in range(len(keys))}

print(dic)
```

12. Sort a dictionary by its values 
- Input: {'a':3, 'b':1, 'c':2}

```python
dicto = {'a':3, 'b':1, 'c':2}

# Convert dictionary into list of tuples
items = list(dicto.items())   # [('a',3), ('b',1), ('c',2)]

# Bubble sort by values
for i in range(len(items)):
    for j in range(len(items)-i-1):
        if items[j][1] > items[j+1][1]:
            # Swap
            items[j], items[j+1] = items[j+1], items[j]

# Convert back to dictionary
sorted_dicto = dict(items)
print(sorted_dicto)
```

```python
dicto = {'b':2, 'a':1, 'd':4, 'c':3}

# Step 1: keys ko list mein lo
keys = list(dicto.keys())   # ['b','a','d','c']

# Step 2: bubble sort style sorting
for i in range(len(keys)):
    for j in range(len(keys)-i-1):
        if keys[j] > keys[j+1]:   # compare alphabetically
            keys[j], keys[j+1] = keys[j+1], keys[j]

# Step 3: nayi dictionary banao sorted keys ke order mein
sorted_dicto = {}
for k in keys:
    sorted_dicto[k] = dicto[k]

print(sorted_dicto)
```

23. Count number of nested keys in a dictionary 
- Input: {'a':1, 'b':{'x':2,'y':3}} 
- Output: 4 

```python
Input = {'a':1, 'b':{'x':2,'y':3}, 'c':{}}
count = 0

for key, value in Input.items():
    count += 1   # outer key count
    if type(value) == dict:   # agar value dictionary hai
        for subkey in value.keys():
            count += 1   # nested key count

print("Number of Keys in dictionary =", count)
```
