# Exception Handaling -
                - When we feel that an error might occur in this block of code. 
## try - 
        - If we feel like the code of block through error then we use try block.
## except - 
        - When the try block occur then except block will execute.
## else - 
        - If try block run successfully without execept block. then else block will run.
## finally - 
            - It always run not dependent on try or except block.
