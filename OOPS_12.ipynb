```python
class shap_old:
    def area(self):
        print("Welcome to the area calculator olddd...")

class shape(shap_old):

    def __init__(self,length,width):
        self.length = length
        self.width = width


class ractangle(shape):
    def __init__(self,length,width):
        super().__init__(length,width)       # self perameter likhanee ki jarurat nahi hoti.
                                            # class ka name yaad nahi karna padata

    def area(self):                         # Method ko bhi call kiya ja sakta hai.
        super().area()

        print(" Area of rectangle = ",self.length*self.width)

class cube(shape):
    def __init__(self,length,width,height):
        super().__init__(length,width)
        self.height = height


    def area(self):
        super().area()
        print(" Area of cube = ",self.length*self.width,self.height)


rect = ractangle(8,12)
rect.area()

cub = cube(9,12,20)
cub.area()

print(cube.mro())
```

```python
class shap_old:
    def area(self):
        print("Welcome to the area calculator olddd...")

class shape(shap_old):

    def __init__(self,length,width):
        self.length = length
        self.width = width

    def area(self):
        print("Welcome to the area calculator -----")    # python apanne aap hi decide kr leta hai ki kon sa method call karna  hai.


class ractangle(shape):                 # class ractangle(shape): left to right chek karega .
    def __init__(self,length,width):
        super().__init__(length,width)


    def area(self):
        super().area()

        print(" Area of rectangle = ",self.length*self.width)


rect = ractangle(8,12)
rect.area()

print(ractangle.mro())   # mro decide karta hai ki sequence kya hoga run kr ke dekhana hai.
                         # Method resolution order.
```

# Abstract Method -
An abstract method is a method that is declared but contain no implementation Subclasses of the abstract class must provide an implementation of the abstract methods.

```python
class ractangle:
    def area(self,length,width):
        print("Area of ractangle is :",length*width)

class square():
    def print_area(self,side):                     # yaha aak hi kaam ko 3 alag alag method use karna pd raha , nikalana sabko area hi hai .
        print("Area of ractangle is :",side*side)

class cube():
    def show_cube(self):
        print("This is cube")

r = ractangle()
r.area(5,7)

s = square()
s.print_area(7)

c = cube()
c.show_cube()

```

```python
from abc import ABC, abstractmethod       # use to access abstracct class & method  . ABC (abstract class )

class shap(ABC):              #  Ye abstract class hai or is class ka koii obj nahi abana sakte..
    @abstractmethod
    def area(self):   # abstract method ye baan gaya finaal meethod jo sb me use hoga .
        pass

    def welcome(self):
        print('Wlecome to area clculator ')   # ye aabstract nahi hai kyuki @aabstractmethod use nahi kiya


class ractangle(shap):
    def area(self,length,width): # agr ham abstract method nahi lete hai to welcome bhi print nahi hoga.
        print("Area of ractangle is :",length*width)



class square(shap):
    def area(self,side):                     # yaha aak hi kaam ko 3 alag alag method use karna pd raha , nikalana sabko area hi hai .
        print("Area of ractangle is :",side*side)




r = ractangle()
r.area(5,7)
r.welcome()


s = square()
s.area(7)

```

```python
from abc import ABC,abstractmethod
class bankapp(ABC):
    def database(self):
        print("Customer data ")

    @abstractmethod
    def security(self):  # aabstractmethod hamesha empty rahti hai. yaha condition ye hai ki 
        pass             # jab tk app security method use nahi karte app customer ke data ko access nahi krr sakte.

class webpage(bankapp):
    def security(self):
        print('Through the security i can access customer data ')
        print(self.database)

employe = webpage()
employe.security()
employe.database()
```
