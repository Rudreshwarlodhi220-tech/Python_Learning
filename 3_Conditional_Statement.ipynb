# Conditinoal Statement - 
- In this statement we check either condition is True then execute further or False jump out of if block.
- Which statement we want to pass .
- if = for passing the condition.
- else = If if statement will False then else will execute .
- elif = for passing multiple condition .

```python
num = int(input("Enter is your number = "))
if num > 0:
    print("Your number is Positive",num)
else:
    print("Your number is Negative",num)
```

## Electricity bill genaration 

```python
# unit<100 = free
# unit > 100 and unit < 200 = 5 per unit
# 200 < unit = 10 per unit
unit = int(input('Enter your unit = '))
per_unit_price = 0
bill_amount = 0

if unit < 100:
    print("Your bill is = ",bill_amount)

elif unit > 100 and unit < 200 :
    bill_amount = (unit - 100) * 5
    print('Your Electricity bill is =', bill_amount)

elif unit > 200 :
    for_10 = (unit - 200) * 10
    for_5 = 100 * 5
    bill_amount = for_10 + for_5
    print('Your Electricity bill is = ',bill_amount)

```

## 1000 >  amount = 0%  discount
## 1000 - 2000 = 10 % discount
## 2000 < amount= 20% discount


```python

amount = int(input("Enter amount = "))

if amount <= 1000:
    print('0 % discount applied ')
    print("Your amount is = ",amount)

elif amount >= 1000 and amount < 2000:
    amount = ((10/100)*amount) + amount
    print('You got 10% discount')
    print('Then your amount is = ',amount)

elif amount >= 2000 :
    amount = ((20/100)*amount)
    print('You got 20% discount and your final amount is = ',amount)
```
